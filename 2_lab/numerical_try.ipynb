```python
import numpy as np
import matplotlib.pyplot as plt

# Define range around x0 = 0
x0 = 0
x = np.linspace(-1, 1, 400)
h = x - x0

# Compute terms (x - x0)^m and (x - x0)^(m+1)
m = 2
term_m = h ** m
term_m1 = h ** (m + 1)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x, term_m, label=r'$(x - x_0)^m$', linewidth=2)
plt.plot(x, term_m1, label=r'$(x - x_0)^{m+1}$', linestyle='--', linewidth=2)
plt.axvline(0, color='gray', linestyle=':', linewidth=1)
plt.title(r'Comparison of $(x - x_0)^m$ and $(x - x_0)^{m+1}$ near $x_0=0$', fontsize=14)
plt.xlabel(r'$x$', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.savefig("img/remaidner.png", dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Ensure folder exists
os.makedirs("img", exist_ok=True)

# Define function and data
f = np.sin
x0 = np.pi / 4
x = np.linspace(x0 - 1.5, x0 + 1.5, 300)
y = f(x)
h = 0.5

# 1st-order Taylor approximation around x0
taylor_approx = f(x0) + np.cos(x0) * (x - x0)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r"$f(x)=\sin(x)$", color="black", linewidth=2)
plt.plot(x, taylor_approx, "--", label="Taylor approx. (1st order)", color="orange", linewidth=2)
plt.scatter([x0], [f(x0)], color="red", s=70, label=r"$x_0$")
plt.title("1st Order Taylor Approximation of $f(x)=\sin(x)$", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()


plt.savefig("img/taylor.png", dpi=300)
plt.show()
```

```python
# Forward difference approximation
fw = (f(x0 + h) - f(x0)) / h

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, f(x), label=r"$f(x)=\sin(x)$", color="black", linewidth=2)
plt.plot(x, f(x0) + fw * (x - x0), '--', label="Forward difference (approx.)", color="blue", linewidth=2)

# Highlight points
plt.scatter([x0, x0 + h], [f(x0), f(x0 + h)], color="red", s=60, zorder=5, label="Sample points")

# Decorate
plt.title("Forward Difference Approximation", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()

# ✅ Save first, then show
plt.savefig("img/fw.png", dpi=300)
plt.show()
```

```python
# Function and parameters
f = np.sin
x0 = np.pi / 4
h = 0.5

# Backward difference approximation
bw = (f(x0) - f(x0 - h)) / h

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, f(x), label=r"$f(x)=\sin(x)$", color="black", linewidth=2)
plt.plot(x, f(x0) + bw * (x - x0), '--', label="Backward difference (approx.)", color="green", linewidth=2)

# Highlight points used
plt.scatter([x0 - h, x0], [f(x0 - h), f(x0)], color="red", s=60, zorder=5, label="Sample points")

# Decorate
plt.title("Backward Difference Approximation", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()

# ✅ Save first, then show
plt.savefig("img/bw.png", dpi=300)
plt.show()
```

```python
# Central difference derivative approximation
cent = (f(x0 + h) - f(x0 - h)) / (2 * h)

# Create plot
plt.figure(figsize=(8, 5))
plt.plot(x, f(x), label=r"$f(x) = \sin(x)$", color="black", linewidth=2)
plt.plot(x, f(x0) + cent * (x - x0), "--", label="Central difference (approx.)", color="orange", linewidth=2)

# Highlight the points used
plt.scatter([x0 - h, x0, x0 + h],
            [f(x0 - h), f(x0), f(x0 + h)],
            color="red", s=60, zorder=5, label="Sample points")

# Decorate
plt.title("Central Difference Approximation of $f'(x)$", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()

# Save and show
plt.savefig(r"img/central_difference.png", dpi=300)  # high-quality save
plt.show()
```
